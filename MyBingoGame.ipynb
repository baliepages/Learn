```python
import random

random_draw_list = random.sample(range(1, 76), 75)

def generate_card():
    card = {
        "B": [],
        "I": [],
        "N": [],
        "G": [],
        "O": [],
    }
    min = 1
    max = 15
    for letter in card:
        card[letter] = random.sample(range(min, max), 5)
        min += 15
        max += 15
        if letter == "N":
            card[letter][2] = "X" # free space!
    return card

def print_card(card):

    for letter in card:
        print(letter, end="\t")
        for number in card[letter]:
            print(number, end="\t")
        print("\n")
    print("\n")

def roll_the_dice():
    dice = {
            "B": [],
            "I": [],
            "N": [],
            "G": [],
            "O": [],
            }
    min = 1
    max = 15
    for letter in ['B', 'I', 'N', 'G', 'O']:
        dice[letter] = random.sample(range(min, max), 5)
        min += 15
        max += 15
        
    randomLetter = random.choice('BINGO')
    randomIndex = random.sample(range(0, 5), 1)[0]
    return randomLetter + '-' + str(dice[randomLetter][randomIndex])
```

```python
players = ['Lonnie', 'Jason', 'Deepika', 'Justin', 'Bailey', 'Justin', 'Khushboo', 'Krutika', 
           'Monica', 'Yoseph', 'Jessica', 'Unni', 'Robyn', 'Shobana']
for player in players:
    print("\nHere is your card, {0}\n".format(player))
    card = generate_card()
    print_card(card)
```
